### Cleaning Data

- Mengidentifikasi masalah dalam proses pengiriman atau pemrosesan pesanan.

```python
print(data['orders']['order_status'].value_counts())
```

```python
delivered_count = data['orders'].order_status.value_counts()['delivered']
total_orders = len(data['orders'])
delivered_percentage = round((delivered_count / total_orders) * 100, 2)

print(f'data yang diperoleh {delivered_percentage}% .')
```

```python
undelivered = data['orders']['order_id'][data['orders']['order_status'] != 'delivered'].values

for key, value in data.items():
    if 'order_id' in value.columns:
        value = value[~value['order_id'].apply(lambda x: x in undelivered)]
        data[key] = value
```

- Mengecek nans

```python
for key, value in data.items():
    for col in value.columns:

        nan_count = value[col].isnull().sum()
        if nan_count > 0:
            nan_percentage = (nan_count / len(value)) * 100
            print(f"{col} (%): {nan_percentage:.2f}")
```

- hanya 8 order yang missing

```python
missing_dates_count = data['orders'].order_delivered_customer_date.isnull().sum()
print(f'{missing_dates_count}')
```

```python
null_deliveries = data['orders']['order_id'][data['orders']['order_delivered_customer_date'].isnull()].values


for key, value in data.items():
    if 'order_id' in value.columns:

        value = value[~value['order_id'].apply(lambda x: x in null_deliveries)]
        data[key] = value
```

```python
data['orders']['delivered_on_time'] = np.where(data['orders']['order_delivered_customer_date'] < data['orders']['order_estimated_delivery_date'], 'On Time', 'Late')
```

```python
print(f'Review columns: {data["reviews"].columns.values}')
data['reviews'].drop(columns=['review_comment_title'], inplace=True)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:  Kategori produk mana yang menghasilkan pendapatan tertinggi?


```python
items_product = data['items'].merge(data['products'], on='product_id', how='inner')
orders_ip = data['orders'].merge(items_product, on='order_id', how='inner')

product_revenue = orders_ip.pivot_table(
    index='product_category_name',
    aggfunc={
        'order_item_id': 'sum',
        'price': 'mean'
    }
)

product_revenue['total'] = product_revenue['order_item_id'] * product_revenue['price']
product_revenue.rename(columns={'order_item_id': 'sell_probability'}, inplace=True)
product_revenue['sell_probability'] /= len(product_revenue)
product_revenue = product_revenue.sort_values(by='total', ascending=False)

top_n = 10
top_products = product_revenue.head(top_n)

bar_colors = ['#D3D3D3'] * len(top_products)

max_index = top_products['total'].idxmax()
bar_colors = ['#D3D3D3'] * (len(top_products) - 1) + ['#90CAF9']

top_products = top_products.sort_values(by='total', ascending=True)

plt.figure(figsize=(10, 6))
plt.barh(top_products.index.astype(str), top_products['total'], color=bar_colors)
plt.xlabel('Total Revenue', fontsize=12)
plt.ylabel('Product ID', fontsize=12)
plt.title(f'Top {top_n} Products by Total Revenue', fontsize=16)
plt.xticks(rotation=45)
plt.grid(axis='x')

plt.tight_layout()
plt.savefig('/content/bestseller_products.png', dpi=300)
plt.show()
```

### Pertanyaan 2: Bagaimana performa penjualan platform E-Commerce selama beroperasi?

```python
all_data = all_data[(all_data["order_approved_at"] >= "2016-01-01") &
                    (all_data["order_approved_at"] <= "2018-12-31")]

monthly_df = all_data.resample(rule='ME', on='order_approved_at').agg({
    "order_id": "nunique",
})
monthly_df.index = monthly_df.index.strftime('%B')
monthly_df = monthly_df.reset_index()
monthly_df.rename(columns={
    "order_id": "order_count",
}, inplace=True)
```

```python
monthly_df = monthly_df.sort_values('order_count').drop_duplicates('order_approved_at', keep='last')
```

```python
monthly_df.head()
```

```python
monthly_df.sort_values(by='order_count')
```

```python
month_mapping = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12
}

monthly_df["month_numeric"] = monthly_df["order_approved_at"].map(month_mapping)
monthly_df = monthly_df.sort_values("month_numeric")
monthly_df = monthly_df.drop("month_numeric", axis=1)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(
    monthly_df["order_approved_at"],
    monthly_df["order_count"],
    marker='o',
    linewidth=2,
    color="#068DA9"
)
plt.title("Number of Orders per Month (2016-2018)", loc="center", fontsize=20)
plt.xticks(fontsize=10, rotation=25)
plt.yticks(fontsize=10)
plt.show()
```

```python
data_2017 = all_data[(all_data["order_approved_at"] >= "2017-01-01") &
                     (all_data["order_approved_at"] <= "2017-12-31")]

monthly_df = data_2017.resample(rule='ME', on='order_approved_at').agg({
    "order_id": "nunique",
})
monthly_df.index = monthly_df.index.strftime('%B')
monthly_df = monthly_df.reset_index()
monthly_df.rename(columns={
    "order_id": "order_count",
}, inplace=True)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(
    monthly_df["order_approved_at"],
    monthly_df["order_count"],
    marker='o',
    linewidth=2,
    color="#068DA9"
)
plt.title("Number of Orders per Month (2017)", loc="center", fontsize=20)
plt.xticks(fontsize=10, rotation=25)
plt.yticks(fontsize=10)
plt.show()
```

```python
data_2016 = all_data[(all_data["order_approved_at"] >= "2016-01-01") &
                     (all_data["order_approved_at"] <= "2016-12-31")]

monthly_df = data_2016.resample(rule='ME', on='order_approved_at').agg({
    "order_id": "nunique",
})
monthly_df.index = monthly_df.index.strftime('%B')
monthly_df = monthly_df.reset_index()
monthly_df.rename(columns={
    "order_id": "order_count",
}, inplace=True)
```
